```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("data.csv")
```

```python
df.head()
```

```python
x=df['Credit amount'].values
y=df['Duration'].values
plt.title("Data",color='g')
plt.xlabel("Credit amount",color='r')
plt.ylabel("Duration",color='r')
plt.plot(x,y,color='y')
plt.grid()
plt.show()
```

```python
x=df['Credit amount'].values

plt.title("Data",color='g')
plt.pie(x)
plt.grid()
plt.show()
```

```python
x=df['Job'].values
y=df['Age'].values
plt.title("Data",color='g')
plt.xlabel("Job",color='r')
plt.ylabel("Age",color='r')
plt.bar(x,y,color='r')
plt.grid()
plt.show()
```

```python
x=df['Job'].values
y=df['Age'].values
plt.title("Data",color='g')
plt.xlabel("Job",color='r')
plt.ylabel("Age",color='r')
plt.stackplot(x,y)
plt.grid()
plt.show()
```
