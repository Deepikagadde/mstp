```python
import matplotlib.pyplot as plt
from matplotlib import style
plt.style.available
y=[1,4,9,16,25]
x1=[1,2,3,4,5]
x2=[1,4,8,12,25]
fig=plt.figure()
a=fig.add_axes([0,0,1,1])
a.set_title('sample graph')
a.set_xlabel('distance')
a.set_ylabel('time')
l1=a.plot(x1,y,color='g')
l2=a.plot(x2,y,color='r')
plt.legend(labels=('x1','x2'),loc='lower left')
style.use('dark_background')
plt.show()
```
